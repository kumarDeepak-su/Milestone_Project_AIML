```python
import pygmt
```

```python
fig = pygmt.Figure()
fig.coast(
    region=[79, 81, 29, 31],
    projection="M8c",
    land="grey",
    water="lightblue",
    shorelines=True,
    frame="a",
)

# store focal mechanisms parameters in a dict
focal_mechanism = dict(strike=23, dip=35, rake=175, magnitude=5)
focal_mechanism = dict(strike=140, dip=74, rake=94, magnitude=5)
focal_mechanism = dict(strike=108, dip=85, rake=99, magnitude=4)
focal_mechanism = dict(strike=32, dip=17, rake=8, magnitude=4)
focal_mechanism = dict(strike=254, dip=87, rake=-91, magnitude=4)
focal_mechanism = dict(strike=269, dip=11, rake=18, magnitude=4)
focal_mechanism = dict(strike=91, dip=86, rake=23, magnitude=5)

# pass the focal mechanism data to meca in addition to the scale and event
# location
fig.meca(focal_mechanism, scale="1c", longitude= 80.745, latitude=29.699, depth=19.8)
fig.meca(focal_mechanism, scale="1c", longitude= 80.579, latitude=29.669, depth=14.4)
fig.meca(focal_mechanism, scale="1c", longitude= 81.538, latitude=29.588, depth=40.1)
fig.meca(focal_mechanism, scale="1c", longitude= 79.433, latitude=30.430, depth=45.4)
fig.meca(focal_mechanism, scale="1c", longitude= 80.720, latitude=29.681, depth=9.1)
fig.meca(focal_mechanism, scale="1c", longitude= 80.129, latitude=30.061, depth=25.1)
fig.meca(focal_mechanism, scale="1c", longitude= 79.438, latitude= 30.551, depth=9.7)
fig.savefig("my-figure.pdf")
fig.show()
```
